# RAM

Data structures are a way to store data in an efficient manner inside of a computer component called Random Access Memory (RAM). RAM is often just referred to as memory.

An array is an ordered collection of contiguous elements, for example [1, 3, 5]. Each of [1, 3, 5] has a binary representation, meaning they can be represented as 0s and 1s. If we use a single byte to store each, then the integer 1 can be represented as 00000001, 3 as 00000011, and 5 as 00000101. (A byte of RAM is made up of exactly 8 bits.)

When we store an array in RAM, each element is stored continously in order - this means elements are side by side

**Integers commonly occupy 4 bytes (32 bits) in memory.** An address and a value gets associated with an integer upon storing it in RAM. An address is just a distinct location that each one of the values is stored at. Each value is stored contiguously in the RAM, just like an array.




*Each int occupies 4 bytes of space, hence addresses are 4 bytes apart*



*Each character occupies 1 byte of space, hence the addresses are 1 byte apart.*
