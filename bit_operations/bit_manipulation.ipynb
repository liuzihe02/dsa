# Bit Manipulation

## Truth Tables and Bit Operations

1. AND `&` - both bits must be 1 to get a 1
2. OR `|` - at least one of bits must be 1 to get a 1
3. XOR `^` - return 1 only if one of 2 bits is a 1
4. Negation `~` - negation of 0 is 1, and negation of 1 is 0
5. Shifting Bits `>>` right shift and `<<` left shift - takes all the bits and shifts them by one position either left or right.
- e.g. `100<<1=000`



> note we can use the python `bin` function to get binary representations of integers

### Problem: Counting Bits

Let's say that we are asked to count the number of 1 bits that are in the binary representation of the integer 23.

To count the number of 1s, we can keep shifting our bits to the right until we get all 0s. We need a way to detect that our current bit is a 1. Recall that 1 & 1 = 1. So, if our current bit,n, results in n & 1 == 1, we can increment our count.

```python
# Counting Bits
def countBits(n):
    count = 0
    while n > 0:
        print(n)
        #if n is odd, will give 1
        #if n is even, will give 0
        if n & 1 == 1:
            count += 1
        #floor division by 2
        n = n >> 1 # same as n // 2
    return count
```

### Problem: Reverse Binary Bits

Given a 32-bit unsigned integer `n` of type `int`, reverse the bits of the binary representation of n and return the result in `int` format

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        res = 0
        for i in range(32):
            #get the ith bit, where 31st bit is the leftmost bit, 0th bit is the rightmost bit
            #goes from leftmost bit to rightmost bit
            bit = (n >> i) & 1
            #shifts the bit and puts it in the relevant position
            res += (bit << (31 - i))
        return res
```
