# Recursion (Double Branch)

We solve the Fibonacci sequence problem using two-branch recursion. The series is:

$$
\begin{align*}
1.& \quad F(0) = 0 \text{ and } F(1) = 1 \\[2ex]
2.& \quad F(n) = F(n-1) + F(n-2)
\end{align*}
$$

Part 1 is the base case, and part 2 is the recursive case. The Fibonacci sequence is a classic example of a recursive function:

the recurrence relation(the recursive formula) is:

$$
fib(n)=fib(n−1)+fib(n−2)
$$

## Conceptualization



```python
# Recursive implementation to calculate the n-th Fibonacci number
def fibonacci(n):
    # Base case: n = 0 or 1
    if n <= 1:
        return n

    # Recursive case: fib(n) = fib(n - 1) + fib(n - 2)
    return fibonacci(n - 1) + fibonacci(n - 2)
```

## Time and Space complexity


On the 1st level (0 indexed), there is 1, on the 2nd level, there are 2, then 4, after which we can see a pattern. Each level has the *potential* to hold 2x the nodes of the previous level. Therefore, a reasonable upperbound for the total number of nodes in the tree is $ 1+2+4..2^n$ . This is a geometric series, similar to what we saw in the dynamic array chapter. We know the last term is the dominating term, and the sum of the series is roughly $2^{n+1}-1$

This means the total number of nodes in the tree is $O(2^n)$. Each node itself is a function call and the sum of 2 values, hence the time complexity of the function is $O(2^n)$
