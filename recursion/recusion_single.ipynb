# Recursion (One Branch)

when a function calls itself, usually with a different input. This is known as a recursive function.

Recursive functions have two parts:

- The base case.
- The recursive case.


## One Branch Recusion



```python
# Recursive implementation of n! (n-factorial) calculation
def factorial(n):
    # Base case: n = 0 or 1
    if n <= 1:
        return 1

    # Recursive case: n! = n * (n - 1)!
    return n * factorial(n - 1)
```

Break it down into sub-problems, when reach base case, start returning - move back "up" the recursion tree because now we have to "piece" together the answers to our sub-problems to get to the final solution.

1. The first portion of the code always checks if the base case conditions are fulfilled.

2. The next portion does the code for recursive expansion

3. The final portion does the code for recursive collapse, what you do to the stuff thats recursed

> It is important to note that if there is no base case in recursion, the recursive case would execute forever resulting in a stack overflow error.

## Iteration and Recursion

iterative implmentation can be much simpler than recursive case, but not always. This iterative solution is also more space efficient as it doesnt require the function call stack. recursion will be useful for trees.

--- 
## Time and Space Complexity

Time: $O(n)$
- n calls are being made to the factorial function, where each call is O(1)

Space: $O(n)$
- Recursion operates off an implicit *stack* known as the function call stack. This is how we return from one function call to the previous one. With n recursive calls, there will be n function calls placed on the stack, which results in O(n) space.
